```python
#list(map(float, test_list))
import seaborn as sns
sns.set()


fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2,figsize=(15, 15))
plt.figure()
ax1.set_ylim(0,1)
ax2.set_ylim(0,1)
ax3.set_ylim(0,1)
ax4.set_ylim(0,1)
ax1.set_xticks([1,2,3,4,5,6])
ax2.set_xticks([1,2,3,4,5,6])
ax3.set_xticks([1,2,3,4,5,6])
ax4.set_xticks([1,2,3,4,5,6])

for classif in ['lstm']:
  acc = pd.read_csv(classif + '/Results/Accuracy/acc.csv', index_col='Unnamed: 0')
  
  f = pd.read_csv(classif + '/Results/F1_scores/f1.csv', index_col='Unnamed: 0')
  g = pd.read_csv(classif + '/Results/G_scores/g.csv', index_col='Unnamed: 0')
  mcc = pd.read_csv(classif + '/Results/MCC/mcc.csv', index_col='Unnamed: 0')
  p = pd.read_csv(classif + '/Results/Precision/pres.csv', index_col='Unnamed: 0')

  ax1.set_xticklabels(acc.columns.tolist())
  ax2.set_xticklabels(acc.columns.tolist())
  ax3.set_xticklabels(acc.columns.tolist())
  ax4.set_xticklabels(acc.columns.tolist())

  
  # f
  fm = list()
  for fi in f.iloc[f.index.tolist().index(dataset)]:
    
    fi = list(fi)
    fi[len(fi)//2] = ','
    fi = "".join(fi)
    fi = list(map(float, ast.literal_eval(fi)))
    fm.append(np.array(fi).mean())


  # g
  gm = list()
  for gi in g.iloc[g.index.tolist().index(dataset)]:
    
    
    gi = list(gi)
    gi[len(gi)//2] = ','
    gi = "".join(gi)
    
    gi = list(map(float, ast.literal_eval(gi)))
    gm.append(np.array(gi).mean())

  #p
  pm = list()
  for pi in p.iloc[p.index.tolist().index(dataset)]:
    
    pi = list(pi)
    pi[len(pi)//2] = ','
    pi = "".join(pi)
    pi = list(map(float, ast.literal_eval(pi)))
    pm.append(np.array(pi).mean())
    
 
  ax1.bar([1,2,3,4,5,6],fm)  
  print(gm)
  ax2.bar([1,2,3,4,5,6],gm)  
  

  
```
